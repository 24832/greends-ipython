```python
import sys
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import myFunctions as mf

sys.path.append('C:/Users/johan/greends/greends-ipython/FORGE/JOHANNA')

Current = pd.read_csv('C:/Users/johan/greends/greends-ipython/Climate_hisafe_knmievaluation_daily_from_01-01-1981_to_31-12-2010.csv', skiprows=[0])
Future = pd.read_csv('../../Climate_hisafe_knmircp85_daily_from_01-01-2006_to_31-12-2100.csv', skiprows=[0])

# create additional vpd column
Current["vpd"] = mf.calculate_vpd(Current["tasmax"],Current["tasmin"],Current["hurs"] )
Future["vpd"] = mf.calculate_vpd(Future["tasmax"],Future["tasmin"],Future["hurs"] )

Future = Future.loc[(Future['Year']>= 2031)]

# a) calculate monthly averages
past_monthly_avgs = Current.groupby(['Year', 'Month']).mean()
future_monthly_avgs = Future.groupby(['Year', 'Month']).mean()

y1 = past_monthly_avgs['vpd']
y2 = future_monthly_avgs['vpd']
fig, ax = plt.subplots(1,1)
ax.plot(np.linspace(1981, 2010, 360), y1, color='blue', label='monthly averages 1981 to 2010')
ax.plot(np.linspace(2031, 2100, 840), y2, color='red', label='predicted monthly averages 2031 to 2100')
fig.suptitle('Vapour Pressure Deficit 1981-2010 plus prediction 2031-2100')
plt.xlabel('Month-Year')
plt.ylabel('VPD [mbar]')
plt.legend()
plt.show()
```

```python
# b) calculate monthly average over all considered years
past_total_monthly_avgs = Current.groupby(['Month']).mean()
future_total_monthly_avgs = Future.groupby(['Month']).mean()

# plot a simple graph
y1 = past_total_monthly_avgs['vpd']
y2 = future_total_monthly_avgs['vpd']
fig, ax = plt.subplots()
ax.bar(np.linspace(1,12,12), y1, color='blue', label='monthly average 1981-2010', width=-0.4, align='edge')
ax.bar(np.linspace(1,12,12), y2, color='red', label='predicted monthly average 2031-2100', width=0.4, align='edge')
xticlabels = ['Jan', 'Feb', 'Mar', 'Apr','May','Jun', 'Jul','Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(1, 13, step=1), xticlabels)
plt.title('Vapour Pressure Deficit 1981-2010 vs. prediction 2031-2100')
plt.legend()
plt.xlabel('Month')
plt.ylabel('VPD [mbar]')
plt.show()
```
